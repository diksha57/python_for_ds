Data visualization is a techinque to present the data in pictorial and graphical format.
Three major consideration for data visulization are: Clarity, Accuracy and Efficiency.
There are many new python data visulization libraries are introduced recently such as: Matplotlib,
Seaborn, Vispy, bokeh, Pygal, folium, networkx. 
The matplotlib is the mostly use python library.it is 2D plotting library.



A plot is a graphical represent of data which shows relationship b/w two or the distribution of
data. 

```python
import numpy as np
import pandas as pd
#import matplotlib library
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate random numbers
randomnum = np.random.rand(25)
```

```python
#view them
print(randomnum)
```

```python
#slect the style of the plot 
style.use('ggplot')
#plot the random numberws 
plt.plot(randomnum,'c',label= 'line one',linewidth = 3,linestyle = "-")
#apply alpha(alpha represent transparety)
plt.plot(randomnum,alpha = 0.4)
#x-label
plt.xlabel('range')
#Y-label
plt.ylabel('Number')
#apply annotation
plt.annotate('Max',ha = 'center',va='bottom',xytext=(5,0.9),xy=(2,1.0) ,arrowprops={'facecolor':'green'})
#print the title
plt.title('Line Plot')
plt.show()
```



Subplot is used to display multiple plots in same window.With subplot,you can arrange plots in a regular grid.Layout and Spacing adjustment are two important factor to be considered while creating subplots.

```python
tep_data = [75,76,87,89,98,77,69,68,98,67]
wind_data =[13,15,17,20,21,24,18,23,24,14]
time_hrs = [2,3,4,5,6,7,8,9,10,12]
humidity = [77,76,78,79,67,64,66,69,81,85]
particitation = [20,25,28,31,38,29,36,41,33,24]
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title("Tem Data")
plt.plot(time_hrs,tep_data,color ='b',linestyle='--',linewidth = 2)
plt.subplot(1,2,2)
plt.title("Wind Data")
plt.plot(time_hrs,wind_data,color ='r',linewidth = 2.5)
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,2,1)
plt.title("Tem")
plt.plot(time_hrs,tep_data,color ='b',linestyle='-',linewidth = 2)
plt.subplot(2,2,2)
plt.title("Wind")
plt.plot(time_hrs,wind_data,color ='r',linestyle='--',linewidth = 2.5)
plt.subplot(2,2,3)
plt.title("Humidity")
plt.plot(time_hrs,humidity,color ='g',linestyle='-.',linewidth = 3)
plt.subplot(2,2,4)
plt.title("Particitation")
plt.plot(time_hrs,particitation,color ='m',linestyle=':',linewidth = 3.5)
```
