```python
from sklearn.datasets import load_boston 
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
boston_real_state_data = load_boston()
```

```python
print(boston_real_state_data.DESCR)
```

```python
#define x-axis
x_axis = boston_real_state_data.data
```

```python
#define y-axis 
y_axis = boston_real_state_data.target
```


# Histogram




Histogram is a graphical representation of probablity. Histogram is a kind of bar chart. Using Matplotlib and it's bar chart fetures, you can create histogram.

```python
#plot histogram
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis,bins=50)
plt.xlabel('Price in 1000s USD')
plt.ylabel('Numbe of houses')
plt.show()
```

# scatter plot
A scatter plot is used to graphically display the relationship between variables. However, to control a plot, it is recommended to use scatter() method.


```python
# plot scatter plot
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_real_state_data.data[:,5],boston_real_state_data.target,color='b')
plt.xlabel('Price in 1000s USD')
plt.ylabel('Numbe of houses')
plt.show()
```
